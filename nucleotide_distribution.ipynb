```python
# Your original DNA sequence
dna = "ATGCCGT"

# Function to calculate GC content
def gc_content(sequence):
    sequence = sequence.upper()
    if len(sequence) == 0:
        return 0
    gc = sequence.count("G") + sequence.count("C")
    return (gc / len(sequence)) * 100

# Function to count nucleotide bases
def base_count(sequence):
    sequence = sequence.upper()
    return {
        "A": sequence.count("A"),
        "T": sequence.count("T"),
        "G": sequence.count("G"),
        "C": sequence.count("C")
    }

# Run analysis
gc = gc_content(dna)
counts = base_count(dna)

print("GC Content:", gc, "%")
print("Base Count:", counts)

# --- Visualization ---
import matplotlib.pyplot as plt

plt.pie(counts.values(), labels=counts.keys(), autopct='%1.1f%%', startangle=90)
plt.title("Nucleotide Distribution")
plt.show()
```

def read_sequence(file_path):
    with open(file_path, "r") as f:
        lines = f.readlines()
